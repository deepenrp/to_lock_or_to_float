### Random Forest

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y, shuffle=True, test_size=0.2)
```

```python
# How would a random forest classifier score?
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

### XGBoost

```python
import xgboost as xgb

train = xgb.DMatrix(X_train, label=y_train)
test = xgb.DMatrix(X_test, label=y_test)
```

```python
param = {
    'tree_method': 'exact',
    'booster': 'gbtree',
    'max_depth': 15,
    'min_child_weight': 1,
    'gamma': 0.2,
    'eta': 0.05,
    'objective': 'multi:softmax',
    'num_class': 2} 
epochs = 100
```

```python
model = xgb.train(param, train, epochs)
```

```python
predictions = model.predict(test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)
```

```python
# Predict the decision of the new data point
new_data_dmatrix = xgb.DMatrix(new_data)
pred = model.predict(new_data_dmatrix)
print("Classes are either 0 (Float) or 1 (Lock)")
print(f"The new point was classified as: {pred}")
```
